```python
import yfinance as yf

# Fetch AAPL stock data with a 15-minute timeframe
aapl = yf.Ticker("AAPL")
data = aapl.history(period="5d", interval="15m")
```

```python
data
```

```python
# Calculate VWAP
data['VWAP_given'] = (((data['High'] + data['Low'] + data['Close']) / 3) * data['Volume']).cumsum() / data['Volume'].cumsum()
```

```python
data["ave_price"] = (data['High'] + data['Low'] + data['Close']) / 3
```

```python
data["cum_volume"]= data['Volume'].cumsum()
```

```python
data["vwap_learn"]=data["ave_price"]*data["cum_volume"] / data["cum_volume"]
```
